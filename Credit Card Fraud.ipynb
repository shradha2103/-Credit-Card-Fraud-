```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv("creditcard.csv")
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
data['normAmount']=StandardScaler().fit_transform(data['Amount'].values.reshape(-1,1))
```

```python
x=data.drop(['Time','Class','Amount'],axis=1)
y=data.loc[:,"Class"]
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
frauds=data[data['Class']==1].index.values
actual=data[data['Class']!=1].index.values
```

```python
actual_undersampled=np.random.choice(actual,len(frauds),replace=False)
undersampled_indices=np.concatenate([frauds,actual_undersampled])
np.random.shuffle(undersampled_indices)
undersampled_data=data.iloc[undersampled_indices,:]
```

```python
undersampled_data
```

```python
x_undersample=undersampled_data.drop(['Time','Class','Amount'],axis=1)
y_undersample=undersampled_data.loc[:,"Class"]
```

```python
from sklearn.model_selection import train_test_split
xtrain_u,xtest_u,ytrain_u,ytest_u=train_test_split(x_undersample,y_undersample,test_size=0.3,random_state=0)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
from sklearn.metrics import recall_score,precision_score
```

```python
c=[0.001,0.01,0.1,1,10]
fold=KFold(n_splits=5)
```

```python
for x in c:
    print("--------------------")
    print("value: ",x)
    print("--------------------")
    lr=LogisticRegression(C=x,penalty='l1')
    m=[]
    for train,test in fold.split(y_undersample):
        
        lr.fit(x_undersample.iloc[train],y_undersample.iloc[train].values.ravel())
        y_pred=lr.predict(x_undersample.iloc[test,:])
        print("accuracy score: ",lr.score(x_undersample.iloc[test],y_undersample.iloc[test].values.ravel()))
        rec=recall_score(y_undersample.iloc[test].values.ravel(),y_pred)
        m.append(rec)
        print("recall score: ",rec)
    print("mean recall ",np.mean(m))
```

```python
lr=LogisticRegression(C=0.01,penalty='l1')
lr.fit(xtrain_u,ytrain_u)
```

```python
lr.score(xtest_u,ytest_u)
```

```python
recall_score(lr.predict(xtest_u),ytest_u)
```

```python
lr.score(xtest,ytest)
```

```python
recall_score(lr.predict(xtest),ytest)
```

```python
from sklearn.metrics import precision_score
precision_score(lr.predict(xtest),ytest)
```
